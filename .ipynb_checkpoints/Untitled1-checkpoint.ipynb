```python
import numpy as np
from statsmodels.datasets import grunfeld
from linearmodels import PanelOLS
```

```python
import pandas.stats
```

```python
data = grunfeld.load_pandas().data
data.year = data.year.astype(np.int64)
# MultiIndex, entity - time
data = data.set_index(['firm','year'])
data.head()
```

```python
mod = PanelOLS(data.invest, data[['value','capital']], time_effects=True)
res = mod.fit()
res
```

```python
mod.entity_effects
```

```python
mod.has_constant
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(data.invest, res.fitted_values.fitted_values)
```

```python
R2 = []
for firm in data.index.levels[0].tolist():
    R2.append(r2_score(data.loc[firm]["invest"], res.fitted_values.loc[firm]["fitted_values"]))
```
